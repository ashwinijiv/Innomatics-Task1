### 5) Write a function

```python
def is_leap(year):
    if year%4==0  and year%100!=0 or year%400==0:
        return True
    else:
        return False
    
year = int(input())
print(is_leap(year))
```

### 6) Print Function

```python
n = int(input())

for i in range(1,n+1):
    print(i,end='')
```

### 7) Python If-Else

```python
n = int(input())
if n%2==0 and n>=2 and n<=5:
    print('Not Weird')
elif n%2==0 and n>=6 and n<=20:
    print('Weird')
elif n%2==0 and n>20:
    print('Not Weird')
else:
    print('Weird')
```
